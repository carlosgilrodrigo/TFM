# 3 - Limpieza de datos

En el proceso KDD, tras la selección del conjunto de datos, hay que limpiar dichos datos en lo que llamamos Data Cleaning. La limpieza de datos pertenece al preprocesado de datos y se basa en la transformación de los datos en bruto que se recogieron, en una forma que pueda ser utilizada en el modelado. Normalmente, las técnicas se basan en adición, eliminación o tranformación de los datos.

3.1 - Tratamiento de missing values



Tratamiento de valores desconocidos o perdidos mediante la eliminación de filas con valores perdidos, la imputación de valores perdidos, el uso de algoritmos que admitan valores perdidos o la predicción de valores perdidos.

Los valores nulos o perdidos no proporcionan información para un modelo. Por ello, y teniendo en cuenta el número de atributos diferentes que tiene el conjunto de datos, se van a eliminar las filas que tengan más del 20% de atributos con datos faltantes, ya que dichas filas no estarían lo suficientemente completas como para tenerlas en cuenta.

```python
def determinar_resultado_final(result):
    golesLocal, golesVisitante = result.split('-')
    golesLocal = int(golesLocal)
    golesVisitante = int(golesVisitante)

    if golesLocal > golesVisitante:
        return 'H'
    elif golesLocal < golesVisitante:
        return 'A'
    else:
        return 'D'

df['result_match'] = df['result_full'].apply(determinar_resultado_final)
```

```python
def determinar_resultado_descanso(result):
    golesLocalDescanso, golesVisitanteDescanso = result.split('-')
    golesLocalDescanso = int(golesLocalDescanso)
    golesVisitanteDescanso = int(golesVisitanteDescanso)

    if golesLocalDescanso > golesVisitanteDescanso:
        return 'H'
    elif golesLocalDescanso < golesVisitanteDescanso:
        return 'A'
    else:
        return 'D'

df['result_descanso'] = df['result_ht'].apply(determinar_resultado_descanso)
df
```

```python
from matplotlib.ticker import FuncFormatter

# Define un orden personalizado para las categorías de 'result_match'
custom_order = ['H', 'D', 'A']

# Aplica el orden personalizado a la columna 'result_match'
df['result_match'] = pd.Categorical(df['result_match'], categories=custom_order, ordered=True)

# Agrupa los datos por 'season' y 'result_match' y cuenta las ocurrencias
result_season_counts = df.groupby(['season', 'result_match']).size().unstack()

# Calcula los porcentajes en cada columna
result_season_percentages = result_season_counts.div(result_season_counts.sum(axis=1), axis=0) * 100

# Crea un gráfico de barras agrupadas con barras más grandes
ax = result_season_percentages.plot(kind='bar', figsize=(26, 10), width=0.9)  # Ajusta el tamaño y el ancho de las barras

# Personaliza el gráfico
title = 'Relación entre Season y Result Match\n'  # Título con salto de línea
plt.title(title, fontsize=16, weight='bold')  # Título en negrita con salto de línea
plt.xlabel('Season', fontsize=14, weight='bold')
plt.ylabel('Porcentaje de partidos (%)', fontsize=14, weight='bold')
plt.xticks(rotation=45, fontsize=12, weight='bold')
plt.yticks(range(0, 101, 10), [f'{i}%' for i in range(0, 101, 10)], fontsize=10, weight='bold')  # Añade % al eje y

# Función para formatear las etiquetas con porcentajes encima de las barras
def annotate_bars(ax):
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height() + 1),
                    ha='center', va='center', fontsize=10, weight='bold')

annotate_bars(ax)

# Mueve la leyenda al lado derecho
ax.legend(title='Result Match', fontsize=12, title_fontsize=12, loc='center left', bbox_to_anchor=(1, 0.5))

# Muestra el gráfico
plt.show()
```
